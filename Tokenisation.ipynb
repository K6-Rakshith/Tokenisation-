```python
# Getting dialogs of speakers
dialogue_list = list(df['Dialogue'])
dialogue_list
```

```python
# Cleaning the dialogues with special characters and repetitions
dialogue_list_lower=[]
dialogue_list_final=[]

for lower_d in dialogue_list:
    str_lower=lower_d.lower()
    dialogue_list_lower.append(str_lower)

dialogue_list_lower_clean = [re.sub("[-()\"#/@;:<>{}`+=~|.!?,…]", "", _) for _ in dialogue_list_lower]

for d in dialogue_list_lower_clean:
    str_test=(' '.join(dict.fromkeys(d.split())))
    dialogue_list_final.append(str_test)
```

```python
# Opening the file to read line by line 
text_file = open("conv.txt","r")
```

```python
# Reading the lines 
lines_at_file = text_file.readlines()
```

```python
# Running for loop to go through every line and to append the list 
l = []
for line in lines_at_file:
    if(line == '\n'):
        continue
    l.append(line[:-2])
```

```python
# checking whether the characters are an alphabet or not through for loop 
def removelast(dicti):
    newDict = []
    for i in dicti:
        if(i == ''):
            continue
        str = i.lower()
        if(not(str[-1].isalpha())):
            newDict.append(str[:-1])
        else:
            newDict.append(str)
    return newDict
```

#### Finding unique words of each speaker

```python
# Using Dictionary to store unique characters with new file 
dict = {}
speaker_characters = []
for line in l:
    ch = line.split(': ')[0]
    dicti = line.split(': ')[1]
    s = set(removelast(dicti.split(" ")))
    if(dict.get(ch) == None):
        dict[ch] = s
        speaker_characters.append(ch)
    else:
        updated = dict.get(ch)
        updated.update(s)
        dict.update({ch:updated})
```

```python
# Displaying characters with the new file opened 
speaker_characters
```

```python
# Getting the speakers with their unique characters in dictionary
dict
```

```python
# filtering and cleaning the words to store them in a file
dialogue_list_lower=[]
dialogue_list_final=[]

for lower_d in dialogue_list:
    str_lower=lower_d.lower()
    dialogue_list_lower.append(str_lower)

dialogue_list_lower_clean = [re.sub("[-()\"#/@;:<>{}`+=~|.!?,…]", "", _) for _ in dialogue_list_lower]

for d in dialogue_list_lower_clean:
    str_test=(' '.join(dict.fromkeys(d.split())))
    dialogue_list_final.append(str_test)
```
